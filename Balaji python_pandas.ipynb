```python
# Calcular la probabilidad de "Cash" y "Online" para cada item_name

# Calcular el porcentaje de cada 'transaction_type' para cada 'item_name'
probabilidades = df.groupby('item_name')['transaction_type'].value_counts(normalize=True).unstack(fill_value=0) * 100

# Mostrar las probabilidades para cada producto
print(probabilidades)
```

```python
import numpy as np

# Definir una función para reemplazar nulos basándonos en las probabilidades
def reemplazar_nulos(row, probabilidades):
    if pd.isnull(row['transaction_type']):
        item_name = row['item_name']
        
        # Probabilidades de Cash y Online para el producto
        prob_cash = probabilidades.loc[item_name, 'Cash']
        prob_online = probabilidades.loc[item_name, 'Online']
        
        # Crear un valor aleatorio para elegir entre Cash o Online
        valor_aleatorio = np.random.uniform(0, 100)
        
        # Reemplazar con 'Cash' o 'Online' según el valor aleatorio
        if valor_aleatorio < prob_cash:
            return 'Cash'
        else:
            return 'Online'
    else:
        return row['transaction_type']

# Aplicar la función a todas las filas del DataFrame
df['transaction_type'] = df.apply(reemplazar_nulos, axis=1, probabilidades=probabilidades)

# Mostrar el DataFrame con los valores nulos reemplazados
df.head(10)
```

```python
# Extraer el nombre del día de la semana
df['day_of_week'] = df['date'].dt.day_name()

# Agrupar por día de la semana y sumar las ventas (o la columna de interés)
ventas_por_dia = df.groupby('day_of_week')['transaction_amount'].sum()

# Ordenar los resultados para encontrar el día con más ventas
ventas_por_dia_ordenadas = ventas_por_dia.sort_values(ascending=False)

# Mostrar el día con más ventas
ventas_por_dia_ordenadas
```

```python
# Sumar las ventas por item_type y time_of_sale
ventas_por_tipo_y_hora = df.groupby(['item_type', 'time_of_sale'])['transaction_amount'].sum().reset_index()

# Mostrar la suma de las ventas por item_type y time_of_sale
print(ventas_por_tipo_y_hora)

# Calcular el ticket promedio por item_type
# El ticket promedio se calcula como la suma total de transaction_amount por tipo dividido por la cantidad total de productos vendidos
ticket_promedio_por_tipo = df.groupby('item_type').apply(lambda x: x['transaction_amount'].sum() / x['quantity'].sum()).reset_index(name='ticket_promedio')

# Mostrar el ticket promedio por tipo de producto
print(ticket_promedio_por_tipo)
```

```python
# Extraer el mes y el año de la columna 'date'
df['mes_ano'] = df['date'].dt.to_period('M')

# Agrupar por la columna 'mes_ano' y sumar las ventas totales (transaction_amount)
ventas_por_mes = df.groupby('mes_ano')['transaction_amount'].sum().reset_index()

# Mostrar las ventas totales por mes
print(ventas_por_mes)
```
